### Cleaning

```python
backup = car.copy()
```

```python
car = car[car['year'].str.isnumeric()]
```

```python
car['year'] = car['year'].astype('int32')
```

```python
car = car[car['Price'] != 'Ask For Price']
car
```

```python
car['Price'] =car['Price'].str.replace(',','').astype('int')
```

```python
car['kms_driven'] = car['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
car = car[car['kms_driven'].str.isnumeric()]
```

```python
car['kms_driven'] = car['kms_driven'].astype('int')
```

```python
car.info()
```

```python
car['fuel_type'].isna().fillna('Petrol',inplace=True)
```

```python
car['fuel_type'] = car['fuel_type'].fillna('Petrol')
```

```python
car['name'] = car['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
car
```

```python
car.reset_index(drop=True,inplace=True)
```

```python
car
```

```python
#removing outlier
car = car[car['Price']<6000000].reset_index(drop=True)
```

```python
car.to_csv('cleaned car.csv')
```

```python
x = car.drop(columns='Price')
y = car['Price']
```

```python
x,y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                        remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
ypred = pipe.predict(x_test)
ypred
```

```python
r2_score(y_test,ypred)
```

```python
#for the max r2 score
scores = []
for i in range(1000):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=i)
    lr = LinearRegression()
    pipe = make_pipeline(column_trans,lr)
    pipe.fit(x_train,y_train)
    ypred= pipe.predict(x_test)
    scores.append(r2_score(y_test,ypred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=np.argmax(scores))
lr = LinearRegression()
pipe = make_pipeline(column_trans,lr)
pipe.fit(x_train,y_train)
ypred= pipe.predict(x_test)
scores.append(r2_score(y_test,ypred))

r2_score(y_test,ypred) 
```
